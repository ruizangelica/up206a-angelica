When looking at the latitude and longitude, I notice that they are defined as objects even though they should be floats. Thus, before I begin analysis and plotting the data I need to change both into floats.

```python
# convert site_latitude and site_longitude to float
air_quality['SITE_LATITUDE'] = air_quality['SITE_LATITUDE'].astype(float)
air_quality['SITE_LONGITUDE']= air_quality['SITE_LONGITUDE'].astype(float)
```

```python
# to double check that it changed
air_quality.dtypes
```

I can further analyze the data by examining how much air quality data each site gathered over the course of a year.

```python
# Names of sites and the amount of air quality data gathered for the year (per site)
air_quality['Site Name'].value_counts()
```

I can plot this information, but first I need to convert it into a dataframe.

```python
# save as a variable
site_count = air_quality['Site Name'].value_counts()
site_count
```

```python
# convert to dataframe
site_count = site_count.reset_index()
site_count
```

```python
# to rename columns
site_count.columns = ['site', 'aq gathered']
```

```python
# plot the amount of air quailty data gathered by each site
site_count.plot.bar(x = 'site', y = 'aq gathered', title = 'Amount of Air Quality Data Gathered, by Site')
```

At first glance, it shows that the Los Angeles-North Main Street has the largest amount of PM 2.5 data gathered. Although there are only 365 days in a year, this site shows to have collected data about 800 times. Further analysis would be needed to understand why that number is almost triple the number of days in a year. It could be due to the fact that this site has multiple air monitors or it may collect data more than once a day.

If I wanted to take a deeper dive into the data and only look at a subset of columns, I could run a command to filter my dataframe. 

```python
# list of desired column names
desired_aq_columns = ['Site Name', 'Daily Mean PM2.5 Concentration', 'DAILY_AQI_VALUE', 'SITE_LATITUDE', 'SITE_LONGITUDE']

# to preserve my new dataframe
aq_trimmed = air_quality[desired_aq_columns].copy()

#to rename columns
aq_trimmed.columns = ['site', 'daily_PM', 'DAILY_AQI_VALUE', 'SITE_LATITUDE', 'SITE_LONGITUDE']

aq_trimmed
```

Since the Los Angeles-North Main Street site had the most counts per year, I can filter the data in order to only look at that column.

```python
# filter the data
aq_trimmed[aq_trimmed.site == 'Los Angeles-North Main Street']
```
